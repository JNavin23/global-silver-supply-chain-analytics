DATA CLEANING

```python
print("Text Inconsistencies Fixed")
silver['mine_country'] = silver['mine_country'].str.title()
```

```python
print("Zero Quantity Fixed")
silver.loc[silver['quantity_kg'] == 0, 'quantity_kg'] = np.nan
```

```python
print("Missing Price Values Fixed")
silver['price_per_kg'] = silver.groupby('silver_quality')['price_per_kg'].transform(lambda x: x.fillna(x.median()))
```

```python
print("Delivery Delay Nulls Fixed")
silver['delivery_delay_days'] = silver['delivery_delay_days'].fillna(0)
```

```python
print("Operator Notes Fixed")
silver['operator_notes'] = silver['operator_notes'].fillna("All Fine")
```

FEATURE ENGINEERING

```python
print("Revenue Column Created")
silver['revenue'] = silver['quantity_kg'] * silver['price_per_kg']
```

```python
print("Delay Flag")
silver['is_delayed'] = np.where(silver['delivery_delay_days'] > 0, "Yes", "No")
```

```python
#Data Features
silver['shipment_date'] = pd.to_datetime(silver['shipment_date'])
silver['year'] = silver['shipment_date'].dt.year
silver['month'] = silver['shipment_date'].dt.month
silver['month_name'] = silver['shipment_date'].dt.month_name()
```
